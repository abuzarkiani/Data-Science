# Analysis

# Importing Libraries

```python
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt
```

```python
data =pd.read_csv('data.csv')

df = data.copy()

df.head()
```

# Column Names

```python
df.columns.unique()
```

# Removing Unnecessary Columns

```python
df.drop(['cname','cedition','cproddat','cseqno','name','essround','edition','idno', 'dweight', 'pspwght', 'pweight'],inplace=True,axis=1)
```

```python
df.head()
```

# Rename Columns Name

```python
df.rename(
    columns={
        "cntry": "Country",
        "tvtot": "Watching_TV",
        "tvpol": "Watching_news_on_TV",
        "rdtot": "Listening_Radio",
        "rdpol": "Listening_news_on_Radio",
        "nwsptot": "Newspaper_Reading",
        "nwsppol": "Reading_news",
        "netuse": "Using_internet",
        "ppltrst": "Trust",
        "pplfair": "Fairness",
        "pplhlp": "Help"
        
    },
    inplace=True
)
```

```python
df.head()
```

# Visualizing missing values

```python
msno.matrix(df)  
```

# Count of Null Values

```python
df.isnull().sum()
```

# Dropping Null Values

```python
df = df.dropna()
df.head()
```

# Creating Country Categorical Variable

```python
df["Country"] = df.Country.map({'DK' : "Denmark",
                                  'FI' : "Finland",
                                  'NO' : "Norway",
                                  'SE' : "Sweden"
                                 })
```

# Data Types

```python
df.dtypes
```

# Questions

# 1. Country wise trend of TV watching culture in Nordic States

```python
country_wise_tv_table = pd.crosstab(index=df["Country"], 
                          columns=df["Watching_TV"])

country_wise_tv_table
```

```python
country_wise_tv_table.plot(kind="bar", figsize=(25,20),stacked=True,color={'tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan'})
```

# 2. Linear and Non-Linear media consumption trends in Nordic countries

```python
country_wise_internet_table = pd.crosstab(index=df["Country"], 
                          columns=df["Using_internet"])

country_wise_internet_table
```

```python
country_wise_internet_table.plot(kind="bar", figsize=(25,20),stacked=True,color={'tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan'}).plot(kind="bar", figsize=(13,13),stacked=True,color={'tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan'})
```

# 3. Tendency to trust others in Nordic Nations

```python
country_wise_trust_table = pd.crosstab(index=df["Country"], 
                          columns=df["Trust"])

country_wise_trust_table
```

```python
country_wise_trust_table.plot(kind="bar", figsize=(25,20),stacked=True,color={'tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan'}).plot(kind="bar", figsize=(13,13),stacked=True,color={'tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple', 'tab:brown', 'tab:pink', 'tab:gray', 'tab:olive', 'tab:cyan'})
```
